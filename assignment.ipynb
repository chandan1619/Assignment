# downloaded tom sayer dataset
# tom is our data set
# counted the frequency of all words in the tom sayer set
# claculated type ,token and their ratio


```python
import nltk as n;
import re;
import operator

f=open("tom",'r')
rd=f.read();
x=list(n.sent_tokenize(rd));

d={};
tokens=0;
for i in x:
    wordlist=n.word_tokenize(i);
    for word in wordlist:
        if  re.match("[a-z]+",word):
            tokens+=1;
            if(word not in d):
                d[word]=1;
            else:
                d[word]=d[word]+1;

d1 = dict(sorted(d.items(), key=operator.itemgetter(1),reverse=True));
    

for keys in d1:
    print(keys,d1[keys])
    
type=len(d1);
print("types=", type, "tokens=",tokens)

print("TTR=",type/tokens)
    
    
    
```

# Heap's law

|V|=k*pow(N,beta)

we taken the log on both side with base e

log(|V|)=log(k)+beta*log(N);

if we assume log(|v|) as y and log(N) as x;

it is similar to
y=mx+c;

where m=beta and c=log(k);


so we plot between log(|v|) and log(N)

and find the slop and intercept of the line which basically givime me the value of beta and k;

for k=inverse  log of c;

```python
from matplotlib import pyplot as plt

import math as m

vocabulary=[];
tokens=[];

d={};

f=open("tom",'r');
i=0;
token=0;
while i<10:
    for j in range(10):
        x=f.readline();
        words=x.split(" ")
        for word in words:
            token+=1;
            if word not in d:
                d[word]=1;
            else:
                d[word]=d[word]+1;
    vocabulary.append(m.log(len(d)));
    tokens.append(m.log(token));
    d.clear();
    token=0;
            
    i+=1;
print(vocabulary);
print(tokens);

plt.title("Heap's law");
plt.xlabel("log of tokens")
plt.ylabel("log of vacabulary")
plt.scatter(tokens,vocabulary)
plt.show();




```

# we have calculate the slope and intercept of line

# for calculating k and beta

```python
from statistics import mean
import numpy as np
import math
def slope_best(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)))
    
    b = mean(ys) - m*mean(xs)
    
    return m, b

# m is slope of line
# b is intercept of line
xs = np.array(tokens)
ys = np.array(vocabulary)
m, b = slope_best(xs,ys)

# taken antilog for calculating k
b1=math.pow(math.e,b);
print("The value of beta=",m);
print("The value of k=",b1);

```
