# downloaded "Alice in Wonderland" text book

# Divided into two file named trainnning.txt and testing.txt in 80 and 20 ratio.

# for identifying the start and end of the sentence i added two unigrams  named sent_start and sent_end before and after each sentence  respectively.




```python
import nltk
f=open('data.txt').read()
lines=nltk.sent_tokenize(f)
count=0;
for line in lines:
    if(line!='\n'):
        count+=1;
count1=count*80//100        
f1= open("trainning.txt", "w")
f2=open("testing.txt","w")
i=0;
for line in lines:
    
    if(i<count1):
        line="sent_start " + line + " sent_end "
        f1.write(line)
    else:
        line="sent_start " + line + " sent_end "
        f2.write(line)
    i+=1;
       
```

# stored the frequency of all unigram in the dictionary and then computed the MLE of all unigrams

```python
import nltk
import re
f3=open('trainning.txt').read()



unigram_freq={};


lines=nltk.sent_tokenize(f3)
for line in lines:
    words=nltk.word_tokenize(line);
    for word in words:
        if(word not in unigram_freq):
            unigram_freq[word]=1;
        else:
            unigram_freq[word]=unigram_freq.get(word)+1;

ulength=len(unigram_freq.keys());

print("Number of unigrams=",ulength)
        
for keys in unigram_freq.keys():
    print(keys,":",unigram_freq[keys]/ulength)



```

# 4(a)  Generator(bigrams) - from the above given text we have taken 1000 bigrams using multinomial function and generated  a duplicate text from that.

```python
import nltk
import random
import numpy as np


f3= open('trainning.txt').read()
words=nltk.word_tokenize(f3);
count=0;
bigram_freq={}
for i in range(len(words)-1):
    s=words[i]+" "+words[i+1]
    count+=1;
    if(s not in bigram_freq):
        bigram_freq[s]=1;
    else:
        bigram_freq[s]=bigram_freq.get(s)+1;

blength=count;
bigram_prob=[]
bigrams=[]

for keys in bigram_freq.keys():
    bigrams.append(keys);
    bigram_prob.append(bigram_freq[keys]/count)
    
    
s=np.random.multinomial(10000,bigram_prob,1000)

output_sentence="";
x=0;
max=-10000;
for row in s:
    max=-1000
    for i in row:
        if(i>max):
            max=i;
    x+=1;
    output_sentence+=bigrams[max]+" ";

print(output_sentence);


        



```

# computed the perplexity value for test data set for the bigram model using good turing smothing since we know the value of constant discounting factor 

```python
import collections
import math
f4=open('testing.txt').read()
words=nltk.word_tokenize(f4)

count=0;
for i in range(len(words)-1):
    s=words[i]+" "+words[i+1]
    count+=1;

prob=1;
perplexity=1;

for i in range(len(words)-1):
    s=words[i]+" "+words[i+1]
    if(s in bigram_freq):
        prob=(bigram_freq.get(s)-0.5)/(unigram_freq.get(words[i],0)+ulength)
        prob=math.pow(prob,1/count)
        perplexity*=prob;

print(1/prob)
    








```
